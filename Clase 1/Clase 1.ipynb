Ej 1: Definir una función leer_parque(nombre_archivo, parque) que abra el

(recordar que cada fila del csv corresponde a un árbol).


```python
parques_ = df['espacio_ve']
parques =  list(set(parques_))

columnas = ['long', 'lat', 'id_arbol', 'altura_tot', 'diametro', 'inclinacio',
       'id_especie', 'nombre_com', 'nombre_cie', 'tipo_folla',
       'ubicacion', 'nombre_fam', 'nombre_gen', 'origen', 'coord_x',
       'coord_y']

def generar_datos_arbol(index,dataframe):
    datos_arbol = []
    for string in columnas:
       datos_arbol.append(dataframe.loc[index][string])
    return datos_arbol

def leer_parque(dataframe,parque):
    lista = []
    contador = 1
    for index in range(51502):
        dictionary = {}
        if df.loc[index]['espacio_ve'] == parque:
            dictionary[str(f"arbol_{contador}")] = generar_datos_arbol(index, dataframe)
            contador = contador + 1
            lista.append(dictionary)

    return lista 

general_paz = leer_parque(df,'GENERAL PAZ')
print(general_paz)
print(len(general_paz))
```

EJ 5: Escribir una función obtener_inclinaciones(lista_arboles, especie)
que, dada una lista como la generada con leer_parque(...) y una especie
de árbol, devuelva una lista con las inclinaciones (columna 'inclinacio') de
los ejemplares de esa especie.


```python
def obtener_inclinaciones(lista_arboles, especie):
    inclinaciones = []
    for i, dato in enumerate(lista_arboles):
        if lista_arboles[i][f"arbol_{i+1}"][7] == especie:
            inclinaciones.append(float(lista_arboles[i][f"arbol_{i+1}"][5]))
    return inclinaciones

o = obtener_inclinaciones(general_paz, "Jacarandá")
print(o)

```

```python
def especie_promedio_mas_inclinada(lista_arboles):

    especies =  especies_parque(lista_arboles)
    inclinaciones_promedio_totales = {}
    for especie in especies:
        inclinaciones  =  obtener_inclinaciones(lista_arboles, especie)
        inclinacion_maxima_promedio =  np.mean(inclinaciones)
        inclinaciones_promedio_totales[especie] = inclinacion_maxima_promedio

    inclinacion_maxima = max(inclinaciones_promedio_totales.values())
    key_maxima = max(inclinaciones_promedio_totales.items(), key=lambda x:x[1])[0]
    inclinacion_maxima_de_especimen = {}
    inclinacion_maxima_de_especimen[key_maxima] = inclinacion_maxima

    return inclinacion_maxima_de_especimen, inclinaciones_promedio_totales


u,o =  especie_promedio_mas_inclinada(general_paz)
u
```
