```python
consultaSQL = """
SELECT COUNT(*) AS respuestas
FROM respuestas
              """

dataframeResultado = dd.sql(consultaSQL).df()
dataframeResultado
```

```python
consultaSQL = """
            SELECT DISTINCT COUNT(Estudiante) AS c
            FROM respuestas
              """

dataframeResultado = dd.sql(consultaSQL).df()
dataframeResultado
```

P4

```python
consultaSQL = """
            SELECT Transporte, COUNT(Transporte) AS Repetido
            FROM respuestas GROUP BY Transporte 
            HAVING Transporte LIKE'C%'
            ORDER BY Transporte ASC
              """

dataframeResultado = dd.sql(consultaSQL).df()
dataframeResultado
```
