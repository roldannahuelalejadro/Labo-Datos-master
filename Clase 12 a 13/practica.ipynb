```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# ConfiguraciÃ³n inicial
list_columns = wine.columns.tolist()
list_columns.remove('pH')
list_columns.remove('type')
list_columns.remove('quality')

columny = 'pH'

# Configurar el tamaÃ±o de la figura (opcional)
plt.figure(figsize=(12, 8))

# Generar los grÃ¡ficos
for i, column in enumerate(list_columns):
    # Crear un DataFrame temporal con las dos columnas
    temp_df = wine[[column, columny]]
    
    # VisualizaciÃ³n con seaborn (jointplot incluye los histogramas)
    g = sns.jointplot(data=temp_df, x=column, y=columny, kind='scatter', 
                      marginal_kws=dict(bins=25, fill=True))
    
    # AÃ±adir lÃ­nea de regresiÃ³n (opcional)
    #sns.regplot(x=column, y=columny, data=wine, scatter_kws={'alpha':0.3}, line_kws={'color':'red'}, ax=g.ax_joint)
    
    # PersonalizaciÃ³n del tÃ­tulo
    #g.fig.suptitle(f'RelaciÃ³n entre {column} y pH', fontsize=16)
    
    # Ajustar espacio entre la figura y el tÃ­tulo
    g.fig.subplots_adjust(top=0.95)
    
    # Mostrar el grÃ¡fico
    plt.show()
```
